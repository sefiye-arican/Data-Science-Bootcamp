# hiyerarşik kümeleme

```python
df.head()
```

```python
# Assuming kumeler contains the cluster labels
df['Kume'] = kumeler  # Create a new column 'Kume' with cluster labels
df.drop('Kume', axis=1, inplace=True)  # Drop the 'Kume' column
```

```python
from scipy.cluster.hierarchy import linkage
hc_complete = linkage(df, 'complete')
hc_average = linkage(df, 'average')
hc_single = linkage(df, 'single')
```

```python
from scipy.cluster.hierarchy import dendrogram
plt.figure(figsize=(15, 10))
plt.title("complete clustering")
plt.xlabel("index")
plt.ylabel("distance")
dendrogram(hc_complete,leaf_font_size=10) #dendrogam grafiği çıkması lazım normalde
```

```python
from scipy.cluster.hierarchy import dendrogram
plt.figure(figsize=(15, 10))
plt.title("complete clustering")
plt.xlabel("index")
plt.ylabel("distance")
dendrogram(hc_single,leaf_font_size=10)
```

```python
from scipy.cluster.hierarchy import dendrogram
plt.figure(figsize=(15, 10))
plt.title("complete clustering")
plt.xlabel("index")
plt.ylabel("distance")
dendrogram(hc_average,leaf_font_size=10)
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=4,
           affinity="euclidean", linkage="ward")
cluster.fit(df)
```

```python
pd.DataFrame({"Eyaletler:":df.index, "Kumeler":cluster.fit_predict()}).head(10)
```
